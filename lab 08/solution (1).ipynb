```python
import matplotlib.pyplot as plt
```

**TASK 1**

```python
students = [f"S{i}" for i in range(1, 21)]
heights = [150,155,160,165,170,168,172,158,161,169,
           173,162,156,167,171,159,174,152,164,166]
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.subplot(1,1,1)
plt.bar(students,heights)
plt.title('Student Heights')
plt.xlabel("Students")
plt.ylabel("Height(cm)")

plt.tight_layout()
plt.show()
```

```python
plt.subplot(1,2,2)
plt.pie(heights,labels=students,autopct="%1.1f%%")
plt.title("Student Height Distribution (PIE CHART)")

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [5, 18, 12, 22, 28]
```

```python
plt.plot(x,y1,color='pink',marker='o',label='Y1 Line')
plt.plot(x,y2,color='gray',marker='o',label='Y2 Line')

plt.title("Two lines on One graph")
plt.xlabel("Amazing X axis")
plt.ylabel("Incredible Y axis")

plt.legend(loc='lower right')

plt.show()
```

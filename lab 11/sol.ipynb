```python
import cv2
import numpy as np

# TASK 1
img = cv2.imread("image.jpg")
if img is None:
    raise FileNotFoundError("image.jpg not found")

cv2.imshow("Original", img)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray", gray)
resized = cv2.resize(img, (300, 300))
cv2.imshow("Resized", resized)
```

```python
# TASK 2
blank = np.zeros((500, 500, 3), dtype=np.uint8)
cv2.rectangle(blank, (50, 50), (300, 300), (0, 255, 0), 3)
cv2.circle(blank, (400, 400), 50, (255, 0, 0), -1)
cv2.imshow("Shapes", blank)
```

```python
# TASK 3
blur = cv2.GaussianBlur(img, (15, 15), 0)
cv2.imshow("Blur", blur)
crop = img[100:300, 200:400]
cv2.imshow("Crop", crop)
```

```python
# TASK 4
img_text = img.copy()
cv2.putText(img_text, "Hello OpenCV", (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
cv2.imshow("Text on Image", img_text)

blank2 = np.zeros((400, 400, 3), dtype=np.uint8)
cv2.putText(blank2, "Blank Image", (30, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
cv2.imshow("Text on Blank", blank2)
```

```python
# TASK 5
gray2 = cv2.imread("image.jpg", 0)
_, thresh = cv2.threshold(gray2, 120, 255, cv2.THRESH_BINARY)
cv2.imshow("Threshold", thresh)

h, w = gray2.shape
center = (w // 2, h // 2)
matrix = cv2.getRotationMatrix2D(center, 60, 1.0)
rotated = cv2.warpAffine(gray2, matrix, (w, h))
cv2.imshow("Rotated", rotated)
```

```python
# TASK 6
img1 = cv2.imread("img1.jpg")
img2 = cv2.imread("img2.jpg")
if img1 is None or img2 is None:
    raise FileNotFoundError("img1.jpg or img2.jpg not found")

img1 = cv2.resize(img1, (500, 500))
img2 = cv2.resize(img2, (500, 500))

blended = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
cv2.imshow("Blended", blended)

gray_blend = cv2.cvtColor(blended, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray Blend", gray_blend)

equalized = cv2.equalizeHist(gray_blend)
cv2.imshow("Equalized", equalized)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
products = pd.read_csv("products.csv")
orders = pd.read_csv("orders.csv")
```

```python
print(products.head())
print(products.tail(10))

print(orders.head())
print(orders.tail(10))
```

```python
merged = orders.merge(products, on="ProductID")
merged['Revenue'] = merged['Quantity'] * merged['Price']
totalRevenue = merged['Revenue'].sum()
print(totalRevenue)
```

```python
sales = merged.groupby('ProductID')['Quantity'].sum().sort_values(ascending=False)
print(sales.head(5))
print(sales.tail(5))
```

```python
topIds = sales.head(5).index
topProducts = products[products["ProductID"].isin(topIds)]
print(topProducts)
```

```python
avg = products.groupby('Category')['Price'].mean()
print(avg)
```

```python
merged['Order Date'] = pd.to_datetime(merged['Order Date'])
daily = merged.groupby(merged['Order Date'].dt.date)['Revenue'].sum()
monthly = merged.groupby(merged['Order Date'].dt.to_period('M'))['Revenue'].sum()
yearly = merged.groupby(merged['Order Date'].dt.year)['Revenue'].sum()

print(daily.idxmax())
print(monthly.idxmax())
print(yearly.idxmax())
```

```python
monthlydf = merged.groupby(merged['Order Date'].dt.to_period('M'))["Revenue"].sum().to_frame()
print(monthlydf)
```
